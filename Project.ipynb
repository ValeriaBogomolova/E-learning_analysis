```python
registration.head()
```

```python
registration.groupby(['code_module','id_student'], as_index=False) \
    .agg({'date_registration':'count'}) \
    .date_registration.describe()
```

```python
#есть такие студенты, которые несколько раз регистрировались на один и тот же предмет, но в разные семестры
#поскльку задание звучит как найти "наиболее популярные по количеству регистраций", то не будем убирать этих студентов из данных
#ведь их повторные регистрации подтверждают популярность курса
```

```python
reg = registration.groupby('code_module', as_index=False) \
    .agg({'date_registration':'count', 'date_unregistration':'count'}) \
    .sort_values('date_registration',ascending=False)
reg.head(3)
```

```python
unreg = registration.groupby('code_module', as_index=False) \
    .agg({'date_registration':'count', 'date_unregistration':'count'}) \
    .sort_values('date_unregistration',ascending=False)
unreg.head(3)
```

```python
#Возможно наэтих курсах наибольший отток, потому что у них наибольшее число регистраций
#Имеет смысл посчитать % оттока от регистраций на курсе и по нему сравнить (относительное значение)
```

```python
unreg['rate']=unreg.date_unregistration/unreg.date_registration
```

```python
unreg.sort_values('rate', ascending=False).head(3)
```

```python
#Самые популярные по регистрации - BBB, FFF, DDD
#С наибольшим оттоком - CCC, DDD, FFF (по относительному значению)
```
